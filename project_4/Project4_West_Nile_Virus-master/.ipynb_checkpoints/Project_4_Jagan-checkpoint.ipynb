### Data Cleaning

```python
# importing essential libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# Data download
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
spray = pd.read_csv('data/spray.csv')
weather = pd.read_csv('data/weather.csv')
```

```python
#Checking for null values in the four datasets provided
print('Checking for null values in the four datasets:\n\nTRAIN_DATA\n{}\n\nTEST_DATA\n{}\n\nWEATHER_DATA\n{}\n\nSPRAY_DATA\n{}'.format(train.isnull().sum(),test.isnull().sum(),weather.isnull().sum(),spray.isnull().sum()))
```

```python
# As per above check on null values, only the spray dataset has null values. Lets look further into the dataset 
# to determine what is to be done with the null values. The spray data is only for 2011 to 2013 whereas the train 
# data is from 2007 to 2013
spray.sort_values('Date')
```

```python
# And It can be confirmed that the null values were only for the time column for 7th Sep 2011. 
(spray[spray.Time.isnull()])[spray['Date']=='2011-09-07']
```

```python
# The time of the spray is not essential,as the train data only indicates if the virus is present for
# a specific day and not time. Therefore we shall proceed to completely drop the time column from spray
spray.drop(['Time'],axis = 1,inplace=True)
spray.isnull().sum()
```

```python
# Now that there is no more null data in our datasets, lets proceed to look deeper into the datasets
train.info()
```

```python
test.info()
```

```python
sns.pairplot(train)
```

```python
#Merging the datasets for cleaning
train_test = pd.concat([train,test],axis = 0)
```

```python
# As the test data does not have NumMosquitos column(Number of Mosquitoes) and the train data does not have id, they
# shall be removed from the merged train_test dataset
train_test.drop(['NumMosquitos'],axis = 1,inplace =True)
```

```python
# Removal of duplicate rows. The majority of duplicates is because there is one record for every 50 mosquitoes, however these
# can be dropped as we are dropping the number of mosquitoes column itself
train_test.drop_duplicates(inplace=True)
```

```python
#  We will be using longitude and latitude as an indication of the location or address thus we will be dropping all 
# other columns relevant to address. This is based on the assumption that for this case study on mosquitoes,
# the details like block and street are insignificant and the exact geographical coordinates is more appropriate
train_test.drop(['Address', 'Block', 'Street', 'AddressNumberAndStreet','AddressAccuracy'],axis = 1, inplace=True)
```

```python
train_test.info()
```

```python
# Creation of dummy columns for the species
train_test = pd.concat([train_test, pd.get_dummies(train_test.Species)], axis=1)
```

```python
# Separation to cleaned train and test sets
train_clean = train_test[train_test.Id.isnull()]
test_clean = train_test[train_test.Id.notnull()]
```

```python
# Removing the Id column from train set as it is only meant for kaggle submission in the test set
train_clean.drop(['Id'],axis =1 ,inplace=True)
train_clean.info()
```

```python
# Removing the target variable WnvPresent from the test dataset
test_clean.drop(['WnvPresent'],axis =1 ,inplace=True)
test_clean.info()
```

```python
train_clean.to_csv('train_clean.csv',index = False)
```

```python
test_clean.to_csv('test_clean.csv',index = False)
```
