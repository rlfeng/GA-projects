# Cleaning & EDA for weather dataset

## Import libraries and data

```python
import pandas as pd
import numpy as np
from statistics import median, mean
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the weather data
weather = pd.read_csv('../data/weather.csv')
```

```python
# Check the weather datatypes
weather.info()
```

## Check for null and missing values

```python
# Check for NULL values
weather.isnull().sum().sum()
```

```python
# "T" stands for trace amount, which is imputed with "0"
# "M" stands for missing data, which is imputed with mean
for columns in weather.columns:
    #print(columns)
    t_values = weather[weather[columns]=="  T"].count()[columns]
    m_values = weather[weather[columns]=="M"].count()[columns]
    if t_values !=0:
        print("{} has {} '  T' values".format(columns, t_values))
    if m_values !=0:
        print("{} has {} 'M' values".format(columns, m_values))
        print("-------------------------------------")
```

```python
# Replace all "M" values as np.nan
weather.replace('M',np.nan,inplace=True)

# Replace all "T" values as 0.00
weather.replace('  T',0.00,inplace=True)
```

```python
# Change the datatypes to float
columns_float = ['SnowFall','PrecipTotal','StnPressure','SeaLevel','AvgSpeed']
weather[columns_float] = weather[columns_float].astype(float)
```

## Drop unnecessary features

We have decided to drop the following features.
1. **Water1** : All of the 2944 values are missing
2. **SnowFall** : Most of its values are zeros (1459 out of 1472)
3. **CodeSum** : it contains qualitative information that are captured in other quantitative features like DewPoint,PercipTotal.
4. **Depth** : All of the 2944 values are missing

```python
weather['Water1'].isnull().sum()
```

```python
weather['SnowFall'].value_counts()
```

```python
weather['CodeSum'].value_counts()
```

```python
weather['Depth'].value_counts()
```

```python
weather.drop(['Water1','SnowFall','CodeSum','Depth'],axis=1,inplace=True)
```

## Impute missing values

```python
weather.isnull().sum()
```

```python
# Impute missing values for Tavg with average Tmax and Tmin values
weather['Tavg']=(weather.Tmax+weather.Tmin)/2

# Impute missing values for Heat and Cool using Tavg and formula from document
for index,rows in weather[weather['Heat'].isnull()].iterrows():
    if rows['Tavg']>65:
        weather['Cool'].iloc[index] = abs(rows['Tavg']-65)
        weather['Heat'].iloc[index] = 0
    else:
        weather['Heat'].iloc[index] = abs(rows['Tavg']-65)
        weather['Cool'].iloc[index] = 0

# Impute missing values in PrecipTotal with mean and impute trace amount with 0.00
weather['PrecipTotal'].fillna(0.00,inplace=True)

# Impute missing WetBulb values with its mean 
weather["WetBulb"] = pd.to_numeric(weather.WetBulb)
weather["WetBulb"] = weather.groupby("Station")['WetBulb'].transform(lambda x: x.fillna(round(x.mean())))

# Impute missing StnPressure values with its mean 
weather["StnPressure"] = weather.groupby("Station")['StnPressure'].transform(lambda x: x.fillna(round(x.mean(),2)))

# Impute missing SeaLevel values with its mean 
weather["SeaLevel"] = weather.groupby("Station")['SeaLevel'].transform(lambda x: x.fillna(round(x.mean(),2)))

# Impute missing AvgSpeed values with its mean 
weather["AvgSpeed"] = weather.groupby("Station")['AvgSpeed'].transform(lambda x: x.fillna(round(x.mean(),1)))

# Impute the "0" value as the entire station has missing values
weather['Depart'].fillna(0,inplace=True)
```

```python
# Check to confirm all NaN values are imputed
weather.isnull().sum()
```

```python
# Convert object columns to integers
columns_int = ['Tavg','Heat','Cool','Depart']
weather[columns_int] = weather[columns_int].astype(int)
```

```python
weather.info()
```

## Feature engineering for daylight hours

```python
# Feature engineering for daylight hours
weather['Sunset'].replace("-","0000",inplace=True)
weather['Sunrise'].replace("-","0000",inplace=True)

weather['Sunrise'] = weather['Sunrise'].astype(int)
weather['Sunset'] = weather['Sunset'].astype(int)

weather['daylight'] = ((weather['Sunset']-weather['Sunrise']))
```

```python
# Explore correlations between weather features
plt.figure(figsize=(5,4))
sns.set(style="white")
mask = np.triu(np.ones_like(weather.corr(), dtype=np.bool))

sns.heatmap(weather.corr(), mask=mask, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},cmap='coolwarm')
```

## Average values between Stations 1 and 2

```python
# Calculate the average between station 1 and 2 for following average columns
average_columns = [col for col in weather.columns if col not in ['Depart','daylight','Sunrise','Sunset','Station']]
average_data = round(weather[average_columns].groupby(['Date']).mean(),2)
```

```python
# For features (e.g. Depart, Daylight) that are not recorded in Station 2, we will take station 1 values
station1_data = weather[weather['Station']==1][['Date','Depart','daylight']]
final_weather = station1_data.join(average_data,how='inner',on='Date')
final_weather.columns = final_weather.columns.map(lambda col: col.lower())
```

```python
final_weather.rename(columns={'dewpoint':'dew_point',
                             'wetbulb':'wet_bulb',
                             'preciptotal':'precip_total',
                             'stnpressure':'stn_pressure',
                             'sealevel':'sea_level',
                             'resultspeed':'result_speed',
                             'resultdir':'result_dir',
                             'avgspeed':'avg_speed'},inplace=True)
```

```python
final_weather.columns
```

```python
final_weather.to_csv('../data/weather_clean.csv',index=False)
```

## EDA for weather

```python
import math
```

```python
def plotting(df, columns, num_cols=2):
    fig = plt.figure(figsize=(6*num_cols,4*round(len(columns)/num_cols)))
    for col_number, col_name in enumerate(columns):   
        fig.add_subplot(math.ceil(len(columns)/num_cols), num_cols, col_number + 1)
        sns.distplot(df[col_name],bins=10)
        plt.ylabel('Count')
    fig.tight_layout()
```

```python
plotting(final_weather, ['tmax', 'tmin', 'tavg', 'dew_point', 'wet_bulb', 'stn_pressure', 'sea_level', 'avg_speed','daylight','precip_total'])
```
