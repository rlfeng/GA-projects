```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score,auc

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from statistics import median, mean
from math import ceil
```

```python
weather=pd.read_csv('./weather.csv/weather.csv')
```

```python
#Drop the following columns from weather Water1: All rows are missing data CodeSum: Contains information on weather conditions which have been captured in other columns SnowFall: Data are only available for station 1 and most of them are zeros (1459 out of 1472).
weather.drop(columns=['Water1','CodeSum','SnowFall'], inplace=True)
```

```python
#Impute missing 'M' and trace 'T' data for Tavg, WetBulb, StnPressure, SeaLevel, AvgSpeed, PrecipTotal

weather['Tavg']=(weather.Tmax+weather.Tmin)/2
weather['Tavg']=weather['Tavg'].apply(np.ceil).astype(int)

weather['WetBulb'].replace('M',np.nan, inplace=True)
weather['WetBulb']=weather['WetBulb'].astype(float)
weather['WetBulb'].replace('NaN',mean(weather['WetBulb']), inplace=True)

weather['StnPressure'].replace('M',np.nan, inplace=True)
weather['StnPressure']=weather['StnPressure'].astype(float)
weather['StnPressure'].replace('NaN',mean(weather['StnPressure']), inplace=True)

weather['SeaLevel'].replace('M',np.nan, inplace=True)
weather['SeaLevel']=weather['SeaLevel'].astype(float)
weather['SeaLevel'].replace('NaN',mean(weather['SeaLevel']), inplace=True)

weather['AvgSpeed'].replace('M',np.nan, inplace=True)
weather['AvgSpeed']=weather['AvgSpeed'].astype(float)
weather['AvgSpeed'].replace('NaN',mean(weather['AvgSpeed']), inplace=True)

weather['PrecipTotal'].replace('  T',0.00, inplace=True)
weather['PrecipTotal'].replace('M',np.nan, inplace=True)
weather['PrecipTotal']=weather['PrecipTotal'].astype(float)
weather['PrecipTotal'].replace('NaN',mean(weather['PrecipTotal']), inplace=True)
```

```python
# Calculate Heat and Cool parameters from Tavg

weather['heat']=weather['Tavg'].apply(lambda x: 0 if x>64 else 65-x)
weather.replace('Heat', 'heat', inplace=True)

weather['cool']=weather['Tavg'].apply(lambda x: 0 if x<65 else x-65)
weather.replace('Cool', 'cool', inplace=True)
```

```python
# Average the common variables in Stations 1 and 2

weather_final=weather.groupby('Date')['Tavg', 'Tmax', 'Tmin', 'DewPoint', 'WetBulb', 'heat', 'cool', 'PrecipTotal', 'StnPressure', 
                                      'SeaLevel','ResultSpeed','ResultDir','AvgSpeed'].mean()
```

```python
# Feature engineering for daylight hours

sun=weather[weather.Station==1][['Date','Sunset','Sunrise']]
sun['time_diff']=sun.apply(lambda x: int(x.Sunset)-int(x.Sunrise),axis=1)
sun['sun_period']=sun.time_diff.map(lambda x:int(str(x)[:2])+int(str(x)[-2:])/60)

weather_final=weather_final.merge(sun, on=['Date'], how='left')
weather_final.drop(columns=['Sunset','Sunrise','time_diff'], inplace=True)
```

```python
# Merge Station 1's Depth and Depart data with weather_final

depth=weather[weather.Station==1][['Date','Depth']]
depth['Depth']=depth['Depth'].astype(int)
weather_final=weather_final.merge(depth, on=['Date'], how='left')

depart=weather[weather.Station==1][['Date','Depart']]
depart['Depart']=depart['Depart'].astype(int)
weather_final=weather_final.merge(depart, on=['Date'], how='left')
```

# EDA on weather

```python
weather_final_corr=weather_final.corr()
mask = np.triu(np.ones_like(weather_final_corr, dtype=np.bool))
sns.heatmap(weather_final_corr, cmap="PiYG",mask=mask)
```

```python
# Convert Date to month, year
weather_final['year']=pd.to_datetime(weather_final['Date']).dt.strftime('%Y')#.astype(int)
weather_final['month']=pd.to_datetime(weather_final['Date']).dt.strftime('%m')#.astype(int)
```

```python
# Weather variation over months
fig=plt.subplots(figsize=(20,15))
plt.subplot(3,2,1),sns.barplot('month','Tavg',data=weather_final,hue='year')
plt.legend().set_visible(False)
plt.subplot(3,2,2),sns.barplot('month','AvgSpeed',data=weather_final,hue='year')
plt.legend().set_visible(False)
plt.subplot(3,2,3),sns.barplot('month','DewPoint',data=weather_final,hue='year')
plt.legend().set_visible(False)
plt.subplot(3,2,4),sns.barplot('month','sun_period',data=weather_final,hue='year')
plt.legend().set_visible(False)
plt.subplot(3,2,5),sns.barplot('month','StnPressure',data=weather_final,hue='year') 
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
plt.subplot(3,2,6),sns.barplot('month','PrecipTotal',data=weather_final,hue='year')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
```
